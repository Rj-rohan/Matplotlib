# Matplotlib

```python
import matplotlib.pyplot as plt

#Basic
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y)
plt.show()
```

# Bar plot

```python
x=["python","Java","C","Javascript"]
y=[20,40,60,80]
plt.xlabel("Programming languages",fontsize=15)
plt.ylabel("Rating",fontsize=15)
plt.title("Rates over programming languages",fontsize=20)
c=["y","g","b","m"]
plt.bar(x,y,width=0.6,color=c,align="edge",edgecolor="r",linewidth=10,linestyle=":",alpha=1,label="Programming") # align = "center" bydefault
plt.yticks([20,40,60,80],rotation=20)
plt.legend()
plt.show()
```

# Double bar graph

```python
x=["python","Java","C","Javascript"]
y=[20,40,60,80]
z=[10,30,50,70]
plt.bar(x,y,width=0.6,color='g',label="popularity")
plt.bar(x,z,width=0.6,color='r',label="popularity1")
plt.legend()
plt.show()
```

# Scatter plot

```python
x=[1,2,3,4]
y=[11,12,13,14]
plt.xlabel("value on x")
plt.ylabel("value on y")
c=['r','g','b','m']
plt.title("Scatter",fontsize=15)
plt.scatter(x,y,color=c,s=[400,500,600,700],alpha=1,marker="*",edgecolor="r",linewidth=5)
plt.colorbar()
plt.show()
```

# Histogram Plot

```python
import numpy as np
# x=list(np.arange(1,20))
x=[10,62,30,48,53,32,40,61,28,51,21,56,35,57,24,15,25]
plt.hist(x,edgecolor='r',bins=[10,15,20,30,35,40,45,60])
plt.show()
```

# Stem Plot

```python
x=[1,2,4,5,7,3]
y=[3,6,2,6,4,6]
plt.stem(x,y,linefmt=":",markerfmt="r*",bottom=2,basefmt="y",label="python",orientation="horizontal",)
# plt.xticks([3,6])
# plt.yticks([2,4,6,8])
# To increase limits on x and y axis
# plt.xlim(1,10)
# plt.ylim(1,10)
# Axis parameter
plt.axis([0,10,0,10])
plt.legend()
plt.show()
```

```python
x=[1,2,3,4]
y=[3,5,1,6]
plt.plot(x,y)
plt.text(1,2,"Rate1",fontsize=10,style="italic",bbox={"facecolor":"r"})
plt.show()
```

# Subplot

```python
x=[1,2,3,4]
y=[1,2,3,4]
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.pie([1],colors="r")

x1=[10,20,30,40]
plt.subplot(2,2,3)
plt.pie(x1)

plt.show()

```
